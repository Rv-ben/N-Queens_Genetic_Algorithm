```python
import random 
import math
import copy

'''
Input: [[Board],[Board],...] sorted by fitness
Output: [[Board],[Board],...] sorted by survivors
'''
def selection(boards):
    
    #list of selections
    selections = []
    
    #sum of fitnesses
    fitSum = 0
    for x in boards:
        fitSum = fitSum + x.get_fit()
        
    #calculate and store all selections
    for x in boards:
        selections.append(x.get_fit()/fitSum)
    
    #List of indexies for surviving boards
    selectedIndexies = []
    
    # 8 selections
    for i in range(8):
        
        #Random Number 0-1
        r = random.uniform(0, 1)
        
        #Starting range for selection
        startRange  = 0
    
        #Ending range for selection
        endRange = selections[0]
        
        #8 possible ranges
        for j in range(7):
            # StartRange <= r < EndRange 
            if(startRange <= r and r < endRange):
                #Store index of the selected board
                selectedIndexies.append(j)
                break
            if(j==6):
                selectedIndexies.append(j)
            
            #Update the range
            startRange = endRange
            endRange = endRange + selections[i]
    
    survivingBoards = []
    for i in selectedIndexies:
        survivingBoards.append(copy.copy(boards[i]))
    
    return survivingBoards
    
        
```

```python
import random 

'''
input: [[encodedboard],[encodedboard],...] 
output:[[encodedboard],[encodedboard],...]  crossed over
'''

def crossOver(boards):
    
    for i in range(0,len(boards)-1,2):
        #Get a random number 0-7 to split the encodedBoards
        indexSplit = random.randint(1,7)
        
        #Get encodedBoards from both boards
        board1Encode = boards[i].get_encodedBoard()
        
        board2Encode = boards[i+1].get_encodedBoard()
        
        #Set the encodedBoard  = board1(firstHalf) + board2(secondHalf)
        boards[i].set_encodedBoard( board1Encode[:indexSplit]+board2Encode[indexSplit:] ) 
        
        #Set the encodedBoard  = board2(firstHalf) + board1(secondHalf)
        boards[i+1].set_encodedBoard( board2Encode[:indexSplit]+board1Encode[indexSplit:] ) 
```

```python
import random

'''
input: [[Board],[Board],...] non mutated  
output: [[Board],[Board],...] mutated  
'''

def mutation(boards):
    
    #Go through boards
    for board in boards:
        
        #Get random index to mutate 0-8
        mutatedIndex = random.randint(0,8)
        
        #Give a chance to not mutate
        if(mutatedIndex < 8):
            
            #get encodedBoard
            encodedBoard = board.get_encodedBoard()
            
            #mutate the index
            encodedBoard[mutatedIndex] = random.randint(0,7)
            
            #update the board
            board.set_encodedBoard(encodedBoard)
```

```python
def genetic(listB):
    #list B is a list of boards
    
    #Sort the list by fitness
    listB = sorted(listB, key=lambda board:board.fit)
    count = 0
    
    #While the board with best fitness is not equal to 8 choose 2
    while(listB[0].get_fit() != math.comb(8,2)):
        
        #increment count
        count = count + 1
        
        #Use the select function to get surviving boards
        listB = selection(listB)

        #Preform CrossOver
        crossOver(listB)

        #Preform Mutation
        mutation(listB)

        #Update Boards
        for board in listB:
            board.updateBoard()
            
        #Sort Boards by fitness
        listB = sorted(listB, key=lambda board:board.fit)
    
    return [listB[0],count]
```

```python
listB = []

#get 8 random boards in a list
for i in range(8):
    listB.append(Board(8))

print('Computing...')
start = time.time()
winner = genetic(listB)
end = time.time()

print("# of genetic iterations = " + str(winner[1]))   
print("Running time: " + str(end-start) + "s")
winner[0].show()
```
